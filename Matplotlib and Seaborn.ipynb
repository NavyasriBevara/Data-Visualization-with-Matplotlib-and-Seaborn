# Data Visualization with Matplotlib and Seaborn using iris Dataset

### Matplotlib:
> Matplotlib is a popular library for creating static,animated and interactive visualizations.It is widely used in Data science,scientific computing and Data analysis.It also used for plotting data and offers great flexibility and control over the apperance of plots.Matplotlib can produce a variety of plots such as lineplot,Barplot,Histograms and more

### Seaborn:
> Seaborn is a python data visualization library built on top of Matplotlib. It is designed to provide a high-level interface for drawing attractive and informative statistical graphics. Seaborn works seamlessly with pandas data structures and simplifies the process of creating visualizations with just a few lines of code.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading iris Dataset
iris=sns.load_dataset("iris")
```

```python
# Extracting iris data
iris
```

## General Statistics plot

```python
iris.describe()
```

```python
stats_summary=iris.describe().loc[['mean','50%','std']].transpose()
stats_summary.columns=['Mean','Median','Standard Deviation']
```

```python
plt.figure(figsize=(10,6))
stats_summary.plot(kind='bar',color={"skyblue","lightgreen","pink"})
plt.title("General Statistics Summary of Iris Data",fontsize=18)
plt.xlabel("Features",fontsize=15)
plt.ylabel("Values",fontsize=15)
plt.xticks(rotation=360)
plt.tight_layout()
plt.show()
```

## Pie plot for Species Frequency

```python
species_count=iris['species'].value_counts()
```

```python
plt.pie(species_count,labels=species_count.index,startangle=90,autopct='%1.1f%%',colors={'pink','violet','purple'},shadow=True)
plt.title("Species Frequency of Iris Dataset",fontsize=18,loc='center')
plt.show()
```

## Relationship Between Sepal_length and Sepal_width

```python
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris,palette="gist_heat")
plt.title("Sepal_length vs Sepal_width",fontsize=18)
plt.xlabel("Sepal_length",fontsize=15)
plt.ylabel("Sepal_width",fontsize=15)
plt.show()
```

## Distribution of Sepal and Petal Features

```python
sns.pairplot(iris,hue='species',palette='flare')
plt.show()
```

## Jointplot for sepal length and sepal width

> jointplot in seaborn is a powerful tool to visualize the relationship between two variables along with their individual distributions. It combines scatterplots (or other types of plots) with histograms or KDE plots on the axes, which makes it ideal for exploring bivariate relationships.

```python
sns.jointplot(x='sepal_length',y='sepal_width',data=iris,color='purple')
plt.xlabel("Sepal_length",fontsize=15)
plt.ylabel("Sepal_width",fontsize=15)
plt.show()
```

## Kde plot for sepal length and sepal width

> The kdeplot function in seaborn is used to generate Kernel Density Estimate (KDE) plots, which are useful for visualizing the probability density of a continuous variable. KDE plots help in estimating the underlying distribution of data.

```python
setosa=iris[iris['species']=='setosa']
```

```python
sns.kdeplot(x='sepal_length',y='sepal_width',data=setosa,shadow=True,color='purple')
plt.title("Kde plot for sepal length and sepal width",fontsize=17)
plt.xlabel("Sepal_length",fontsize=15)
plt.ylabel("Sepal_width",fontsize=15)
plt.show()
```

## Kde plot for petal length and petal width

```python
sns.kdeplot(x='petal_length',y='petal_width',data=setosa,shadow=True,color='purple')
plt.title("Kde plot for petal length and petal width",fontsize=17)
plt.xlabel("Petal_length",fontsize=15)
plt.ylabel("Petal_width",fontsize=15)
plt.show()
```
